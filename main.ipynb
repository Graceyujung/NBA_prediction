```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(data=games)

plt.title('Box plt for games')
plt.xticks(rotation=45)
plt.xlabel('Attributes')
plt.ylabel('Value')

plt.show()
```

Before we update on cleaned data, i will proceed with robustscaling since there are outliers.

```python
columns_to_drop = [
    'GAME_ID', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'SEASON',
    'TEAM_ID_home', 'TEAM_ID_away', 'GAME_DATE_EST',
    'GAME_STATUS_TEXT', 'PTS_home', 'PTS_away'
]

games = games.drop(columns=columns_to_drop)
games.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Splitting 'games'

X = games.drop(columns=['HOME_TEAM_WINS'])
Y = games['HOME_TEAM_WINS']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
from sklearn.preprocessing import RobustScaler

# Robust Scaling

rs = RobustScaler()
X_scaled = rs.fit_transform(X_train)
X_test_scaled = rs.transform(X_test)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print(Y_train.value_counts())
print(Y_test.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import itertools

# For varied visualization
colormaps = ['Blues', 'Oranges', 'Greens', 'Purples', 'Reds', 'BuGn', 'BuPu', 'coolwarm']
color_cycle = itertools.cycle(colormaps)

def evaluate_model(model, X_scaled, Y_train, X_test_scaled, Y_test):

    model.fit(X_scaled, Y_train)

    y_train_pred = model.predict(X_scaled)
    y_test_pred = model.predict(X_test_scaled)

    # Accuracy
    print("Train Accuracy:", accuracy_score(Y_train, y_train_pred))
    print("Test Accuracy:", accuracy_score(Y_test, y_test_pred))

    train_cmap = next(color_cycle)
    test_cmap = next(color_cycle)

    # Confusion Matrix (Train)
    print("\nConfusion Matrix (Train):")
    sns.heatmap(confusion_matrix(Y_train, y_train_pred), annot=True, fmt='g', cmap=train_cmap)
    plt.title('Train Confusion Matrix')
    plt.show()

    # Confusion Matrix (Test)
    print("\nConfusion Matrix (Test):")
    sns.heatmap(confusion_matrix(Y_test, y_test_pred), annot=True, fmt='g', cmap=test_cmap)
    plt.title('Test Confusion Matrix')
    plt.show()

    # Classification Report (Train)
    print("\nClassification Report (Train):")
    print(classification_report(Y_train, y_train_pred))

    # Classification Report (Test)
    print("\nClassification Report (Test):")
    print(classification_report(Y_test, y_test_pred))

    return model
```

### 1. Logistic Regression

```python
lr = LogisticRegression(class_weight='balanced')
evaluate_model(lr, X_scaled, Y_train, X_test_scaled, Y_test)
```

Challenge: The PTS_home and PTS_away lead to overfitting the model to make the predicted value of the result too high. Since the scores of both teams are equivalent to the game results, this makes the other features unimportant. Therefore, in order to investigate what features to match, PTS need to be removed.

Both Train and Test accuracy are great above 80%, the model has learned well and can generalize to unseen data.

### 2. SVM

```python
from sklearn.svm import SVC

kernels = ['linear', 'poly', 'rbf']
models = {}

for kernel in kernels:
    print(f"Evaluating SVM with kernel: {kernel}")
    svm = SVC(kernel=kernel, probability=True, random_state=42)

    models[f"SVM({kernel})"] = evaluate_model(svm, X_scaled, Y_train, X_test_scaled, Y_test)
```

### 4. KNN

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

def find_optimal_k(X_scaled, Y_train, X_test_scaled, Y_test, k_range=40):
    error_rate = []
    k_values = range(1, k_range)

    for k in k_values:
        knn=KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_scaled, Y_train)
        pred_k = knn.predict(X_test_scaled)
        error_rate.append(1 - accuracy_score(Y_test, pred_k))

    plt.figure(figsize=(10,5))
    plt.plot(k_values, error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=4)
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Error Rate')
    plt.title('Error Rate vs. K Value')
    plt.show()

    optimal_k = k_values[np.argmin(error_rate)]
    print(f"Optimal k found: {optimal_k}")

    return optimal_k
    
```

```python
optimal_k = find_optimal_k(X_scaled, Y_train, X_test_scaled, Y_test, k_range=60)

knn = KNeighborsClassifier(n_neighbors=optimal_k)
evaluate_model(knn, X_scaled, Y_train, X_test_scaled, Y_test)
```

Challenge: from find_optimal_k, with the default k_range=40, the best k value was at 37 (out of 40), so i extended the range to 60 to improve model performance. The graph likely leveled off and showed only minor fluctuations after k=37, indicating the error rate stabilized at that point.

The KNN train and test accuracies before applying the optimal k were 87% and 82%, respectively, indicating the model was overfitting. After applying the optimal k, the accuracies improved and became closer, with both reaching 84%, which indicates a more consistent and balanced performance.

### 6. XGBoost

```python
import xgboost as xgb
import numpy as np

X_scaled_np = np.asarray(X_scaled).astype(np.float32)
X_test_np = np.asarray(X_test_scaled).astype(np.float32)
Y_train_np = np.asarray(Y_train)
Y_test_np = np.asarray(Y_test)

xg = xgb.XGBClassifier(eval_metric='auc')
evaluate_model(xg, X_scaled_np, Y_train_np, X_test_np, Y_test_np)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],  
    'n_estimators': [50, 100, 200, 300],          
    'max_depth': [3, 4, 5, 6, 7],                
    'min_child_weight': [1, 2, 3, 4, 5],           
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],        
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0], 
    'gamma': [0, 0.1, 0.2, 0.3],                    
    'reg_alpha': [0, 0.1, 0.5, 1],                  
    'reg_lambda': [0, 0.1, 0.5, 1]                 
}

rs_xgb = RandomizedSearchCV(estimator=xg, param_distributions=param_dist, 
                                   n_iter=100, cv=5, verbose=2, random_state=42, n_jobs=-1, scoring='accuracy')
rs_xgb.fit(X_scaled_np, Y_train_np)
print("Best parameters found: ", rs_xgb.best_params_)

best_xgb = rs_xgb.best_estimator_

evaluate_model(best_xgb, X_scaled_np, Y_train_np, X_test_np, Y_test_np)
```

Challenge: The dataset is overfitting with XGBoost from a train accuracy of 92.4% and a test accuracy of 83.4%. RandomSearchCV improved the performance, resulting in a train accuracy of 86% and a test accuracy of 84%.

### 7. ROC-AUC Curve

```python
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(model, X_scaled, Y_test, label):

    y_prob = model.predict_proba(X_scaled)[:, 1]

    fpr, tpr, _ = roc_curve(Y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=2, label = f'{label} (AUC = {roc_auc:.2f})')

plt.figure(figsize=(8,6))

models.update({
    "Logistic Regression": lr,
    "Decision Tree": best_dt,
    "KNN": knn,
    "Random Forest": best_rf,
    "XGBoost": best_xgb,
})


for label, model in models.items():
    plot_roc_curve(model, X_test_scaled, Y_test, label)

plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC-AUC) Comparison')
plt.legend(loc="lower right")
plt.show()
```

Excellent performance of SVM(Linear), Logistic Regression and XGBoost with an AUC of 0.93, indicating they are highly effective at distinguishing between classes.

KNN and Random Forest also perform well, with AUCs of 0.92, showing that they are strong contenders.

The Decision Tree has a relatively lower AUC of 0.86, indicating issues like overfitting or a lack of sufficient regularization.
