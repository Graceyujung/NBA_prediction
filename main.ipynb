```python
# Dropping all missing values for now

games.dropna(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(data=games)

plt.title('Box plt for games')
plt.xticks(rotation=45)
plt.xlabel('Attributes')
plt.ylabel('Value')

plt.show()
```

Before we update on cleaned data, i will proceed with robustscaling since there are outliers.

```python
columns_to_drop = [
    'GAME_ID', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'SEASON',
    'TEAM_ID_home', 'TEAM_ID_away', 'GAME_DATE_EST',
    'GAME_STATUS_TEXT', 'PTS_home', 'PTS_away'
]

games = games.drop(columns=columns_to_drop)
games.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Splitting 'games'

X = games.drop(columns=['HOME_TEAM_WINS'])
Y = games['HOME_TEAM_WINS']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
from sklearn.preprocessing import RobustScaler

# Robust Scaling

rs = RobustScaler()
X_scaled = rs.fit_transform(X_train)
X_test_scaled = rs.transform(X_test)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print(Y_train.value_counts())
print(Y_test.value_counts())
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

lr = LogisticRegression(class_weight='balanced')
def evaluate_model(model, X_scaled, Y_train, X_test_scaled, Y_test):

    model.fit(X_scaled, Y_train)

    y_train_pred = model.predict(X_scaled)
    y_test_pred = model.predict(X_test_scaled)

    # Accuracy
    print("Train Accuracy:", accuracy_score(Y_train, y_train_pred))
    print("Test Accuracy:", accuracy_score(Y_test, y_test_pred))

    # Confusion Matrix (Train)
    print("\nConfusion Matrix (Train):")
    sns.heatmap(confusion_matrix(Y_train, y_train_pred), annot=True, fmt='g', cmap='Blues')
    plt.title('Train Confusion Matrix')
    plt.show()

    # Confusion Matrix (Test)
    print("\nConfusion Matrix (Test):")
    sns.heatmap(confusion_matrix(Y_test, y_test_pred), annot=True, fmt='g', cmap='Oranges')
    plt.title('Test Confusion Matrix')
    plt.show()

    # Classification Report (Train)
    print("\nClassification Report (Train):")
    print(classification_report(Y_train, y_train_pred))

    # Classification Report (Test)
    print("\nClassification Report (Test):")
    print(classification_report(Y_test, y_test_pred))

evaluate_model(lr, X_scaled, Y_train, X_test_scaled, Y_test)
```

Challenge: The PTS_home and PTS_away lead to overfitting the model to make the predicted value of the result too high. Since the scores of both teams are equivalent to the game results, this makes the other features unimportant. Therefore, in order to investigate what features to match, PTS need to be removed.

Both Train and Test accuracy are great above 80%, the model has learned well and can generalize to unseen data.

### SVM

```python
from sklearn.svm import SVC

kernels = ['linear', 'poly', 'rbf']
models = {}

for kernel in kernels:
    print(f"Evaluating SVM with kernel: {kernel}")
    svm = SVC(kernel=kernel, probability=True)
    evaluate_model(svm, X_scaled, Y_train, X_test_scaled, Y_test)

    # for roc-auc curve
    models[f"SVM({kernel})"] = svm
```

### KNN

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

def find_optimal_k(X_scaled, Y_train, X_test_scaled, Y_test, k_range=40):
    """Finds the optimal k value by testing different k values and plotting error rates."""
    error_rate = []
    k_values = range(1, k_range)

    for k in k_values:
        knn=KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_scaled, Y_train)
        pred_k = knn.predict(X_test_scaled)
        error_rate.append(1 - accuracy_score(Y_test, pred_k))

    plt.figure(figsize=(10,5))
    plt.plot(k_values, error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=4)
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Error Rate')
    plt.title('Error Rate vs. K Value')
    plt.show()

    optimal_k = k_values[np.argmin(error_rate)]
    print(f"Optimal k found: {optimal_k}")

    return optimal_k
    
# find_optimal_k(X_scaled, Y_train, X_test_scaled, Y_test, k_range=60)
```

```python
optimal_k = find_optimal_k(X_scaled, Y_train, X_test_scaled, Y_test, k_range=60)

knn = KNeighborsClassifier(n_neighbors=optimal_k)
evaluate_model(knn, X_scaled, Y_train, X_test_scaled, Y_test)
```

Challenge: from find_optimal_k, with the default k_range=40, the best k value was at 37 (out of 40), so i extended the range to 60 to improve model performance. The graph likely leveled off and showed only minor fluctuations after k=37, indicating the error rate stabilized at that point.

The KNN train and test accuracies before applying the optimal k were 87% and 82%, respectively, indicating the model was overfitting. After applying the optimal k, the accuracies improved and became closer, with both reaching 84%, which indicates a more consistent and balanced performance.

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
evaluate_model(rf, X_scaled, Y_train, X_test_scaled, Y_test)
```

```python
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(model, X_scaled, Y_test, label):

    y_prob = model.predict_proba(X_scaled)[:, 1]

    fpr, tpr, _ = roc_curve(Y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=2, label = f'{label} (AUC = {roc_auc:.2f})')

plt.figure(figsize=(8,6))

models["Logistic Regression"] = lr
models["KNN"] = knn
models["Random Forest"] = rf


for label, model in models.items():
    plot_roc_curve(model, X_test_scaled, Y_test, label)

plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC-AUC) Comparison')
plt.legend(loc="lower right")
plt.show()
```

Excellent performance of SVM(Linear) and Logistic Regression, suggesting linear models might be well-suited for the dataset. The others provide similar performance with AUC = 0.92, indicating that more complex models are not yielding a significant improvement in this case.
